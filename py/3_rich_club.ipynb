```python
import time
import hypergraph
import rich_club
import matplotlib.pyplot as plt
```

Read hypergraph data.  
This notebook shows an example of randomizing a collaborative grant hypergraph data named *nsf-grants*, which was used in the following paper:  
Kazuki Nakajima, Kazuyuki Shudo, Naoki Masuda. Higher-order rich-club phenomenon in collaborative research grant networks. Scientometrics (2023). [<a href="https://doi.org/10.1007/s11192-022-04621-1">paper</a>]  
This hypergraph is not located in the folder *hyper-dk-series/data/*, although it is privately available upon request.

```python
hypergraph_name = 'nsf-grants'
H = hypergraph.read_hypergraph(hypergraph_name)
```

```python
H.print_info()
```

Compute the normalized rich-club coefficient for the hypergraph.  
Please see the following paper for the definition.  
Kazuki Nakajima, Kazuyuki Shudo, Naoki Masuda. Higher-order rich-club phenomenon in collaborative research grant networks. *Scientometrics* (2023). [<a href="https://doi.org/10.1007/s11192-022-04621-1">paper</a>]

```python
t_s = time.time()
norm_rcc, p_val = rich_club.calc_normalized_rich_club_coefficient(H)
t_e = time.time()
print(t_e - t_s, "[seconds]")
```

'norm_rcc' is a dictionary of the key as node degree *k* and value as the normalized rich-club coefficient for *k*, denoted by $\rho(k)$.

```python
norm_rcc
```

'p_val' is a dictionary of the key as node degree *k* and value as *p*-value derived from the permutation test for $\rho(k)$.

```python
p_val
```

Compute the normalized rich-club coefficient for the subhypergraph induced by the set of hyperedges of given size *s*.  
Specify the set of hyperedge sizes using the variable 'hyperedge_size_set' (in the example below, we focus on hyperedges of sizes two, three, four, and five).  
For the definition, see the following paper:
Kazuki Nakajima, Kazuyuki Shudo, Naoki Masuda. Higher-order rich-club phenomenon in collaborative research grant networks. Scientometrics (2023). [<a href="https://doi.org/10.1007/s11192-022-04621-1">paper</a>]

```python
t_s = time.time()
hyperedge_size_set = {2, 3, 4, 5}
norm_rcc_by_s, p_val_by_s = rich_club.calc_hyperedge_size_dependent_normalized_rich_club_coefficient(H, hyperedge_size_set)
t_e = time.time()
print(t_e - t_s, "[seconds]")
```

'norm_rcc_by_s' is a dictionary of the key as hyperedge size and value as the normalized rich-club coefficient for the subhypergraph induced from the set of hyperedges of size *s*.  

```python
norm_rcc_by_s
```

'p_val' is a dictionary of the key as hyperedge size and value as *p*-value derived from the permutation test for the hypergraph induced from the set of hyperedges of size *s*.  

```python
p_val_by_s
```

```python
# Plot functions
# You can modify these plot functions to suit your needs.

fontsize = 40
plt.rcParams["font.size"] = fontsize
plt.rcParams['xtick.direction'] = 'out' 
plt.rcParams['ytick.direction'] = 'out' 
plt.rcParams['xtick.major.width'] = 2.5 
plt.rcParams['xtick.minor.width'] = 1 
plt.rcParams['ytick.major.width'] = 2.5 
plt.rcParams['ytick.minor.width'] = 1 
plt.rcParams['axes.linewidth'] = 2.5 
plt.rcParams['xtick.major.size'] = 15
plt.rcParams['xtick.minor.size'] = 12.5
plt.rcParams['ytick.major.size'] = 15
plt.rcParams['ytick.minor.size'] = 12.5
plt.rcParams['hatch.linewidth'] = 0.3

# Plot function for the normalized rich-club coefficient for the original hypergraph
def plot_normalized_rich_club_coefficient(norm_rcc):

    xticks = [1, 10, 100,1000]
    xleft = 0.5
    xright = 500
    yticks = [0.6+0.2*i for i in range(0, 4)]
    ybottom = 0.55
    ytop = 1.25

    plt.figure(figsize=(12, 9))
    plt.xscale('log')
    plt.xlabel("Node degree, " + r'$k$')
    plt.ylabel(r"$\rho(k)$")

    x = list(norm_rcc.keys())
    y = list(norm_rcc.values())
    plt.scatter(x, y, color="tomato", marker="o", s=100)

    plt.xticks(xticks)
    plt.yticks(yticks)
    plt.xlim(left=xleft, right=xright)
    plt.ylim(bottom=ybottom, top=ytop)
    plt.tick_params(axis='x', pad=10)
    plt.tick_params(axis='y', pad=10)
    plt.subplots_adjust(left=0.175, right=0.95, bottom=0.175, top=0.95)
    plt.tight_layout()
    plt.show()

    return

# Plot function for the normalized rich-club coefficient for the original hypergraph and the subhypergraph induced from the set of hyperedges of size s. 
def plot_hyperedge_size_dependent_normalized_rich_club_coefficient(norm_rcc, norm_rcc_by_s, hyperedge_size_set):

    xticks = [1, 10, 100,1000]
    xleft = 0.5
    xright = 500
    yticks = [0, 1.0, 2.0, 3.0, 4.0]
    ybottom = 0
    ytop = 4.5
    colors = ["royalblue", "purple", "seagreen", "violet", "purple"]
    markers = ["x", "+", "*", "s"]
    legends = [r"$s = 2$", r"$s = 3$", r"$s = 4$", r"$s = 5$", r"$s = 6$"]

    plt.figure(figsize=(12, 9))
    plt.xscale('log')
    #plt.yscale('log')
    plt.xlabel("Node degree, " + r"$k$")
    plt.ylabel(r"$\rho(k)$")

    x = list(norm_rcc.keys())
    y = list(norm_rcc.values())
    plt.scatter(x, y, color="tomato", marker="o", s=100, label="All hyperedges")
    
    hyperedge_size_list = sorted(list(hyperedge_size_set))

    for i in range(0, len(hyperedge_size_list)):
        s = hyperedge_size_list[i]
        x = list(norm_rcc_by_s[s].keys())
        y = list(norm_rcc_by_s[s].values())
        plt.scatter(x, y, color=colors[i], marker=markers[i], s=100, label=legends[i])

    plt.xticks(xticks)
    plt.yticks(yticks)
    plt.xlim(left=xleft, right=xright)
    plt.ylim(bottom=ybottom, top=ytop)
    plt.tick_params(axis='x', pad=10)
    plt.tick_params(axis='y', pad=10)
    plt.subplots_adjust(left=0.175, right=0.95, bottom=0.175, top=0.95)
    plt.legend(loc="upper right", frameon=False, fontsize=27.5, ncol=1, handletextpad=0.1)
    plt.tight_layout()
    plt.show()

    return
```

Show the normalized rich-club coefficient as a function of node degree k in the hypergraph.

```python
plot_normalized_rich_club_coefficient(norm_rcc)
```

Show the normalized rich-club coefficient as a function of node degree k in the subhypergraph for each *s* specified by the 'hyperedge_size_set'. 
The legend 'All hyperedges' indicates the normalized rich-club coefficient for the original hypergraph.

```python
plot_hyperedge_size_dependent_normalized_rich_club_coefficient(norm_rcc, norm_rcc_by_s, hyperedge_size_set)
```
